# Matplotlib --- Time-series Plot

```python
# This line configures matplotlib to show figures embedded in the notebook, 
# instead of opening a new window for each figure. More about that later. 
# If you are using an old version of IPython, try using '%pylab inline' instead.
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
import matplotlib.dates as mdate
```

```python
x, y = np.loadtxt("inp.dat", unpack=True, converters={0 : mdate.strpdate2num("%Y-%m-%d")})
```

```python
plt.plot_date(x, y, fmt='ro-', xdate=True)
```

```python
fig, ax = plt.subplots()
ax.plot_date(x, y, fmt='ro-', xdate=True)
# update the figure now, otherwise the labels below will be empty...
plt.draw()
# get xticklabels
labels = [xx.get_text() for xx in ax.get_xticklabels()]
# print labels
# Note rotation argument
ax.set_xticklabels(labels, rotation=-45)
# ax.set_xlabel('Date')
# ax.set_ylabel('Value')
# ax.set_title('some title')
plt.show()
```
